```python
import numpy as np
from skimage.transform import resize
from matplotlib import pylab as plt
from matplotlib.colors import ListedColormap
import histomicstk
```

```python
import skimage.io
from PIL import Image
```

```python
# read in images
img1 = skimage.io.imread("./pngs/brain1A.png")
img2 = skimage.io.imread("./pngs/brain1D.png")
```

```python
stain_unmixing_routine_params = {
    'stains': ['hematoxylin', 'eosin'],
    'stain_unmixing_method': 'macenko_pca',
}
tissue_rgb_normalized = histomicstk.preprocessing.color_normalization.deconvolution_based_normalization(
            img1, im_target=img2,
            stain_unmixing_routine_params=stain_unmixing_routine_params)
```

```python
imgplot = plt.imshow(tissue_rgb_normalized)
plt.show()
```

```python
imgplot = plt.imshow(img1)
plt.show()
```

```python
norm_img=Image.fromarray(tissue_rgb_normalized).convert('RGB')
norm_img.save("./pngs/brain1A_normalized.png")
```
